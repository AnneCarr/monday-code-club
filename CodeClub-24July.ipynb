## Task 
- copy your original dataframe
- select the columns you want and remove rows where sex is 'ALL'
- create one dataframe filtered on 30-34 age group
- create a second dataframe filtered on 90-94 age group
- append these two dataframes together (there are quicker ways of doing this - but it's useful to recap)
- make a column which is 10,000's
- plot it

```python
gpAge = gp_data.copy()
```

```python
gpAge = gpAge[['SEX', 'NUMBER_OF_PATIENTS', 'AGE_GROUP_5']]
gpAge.head()
```

```python
gpAge = gpAge[gpAge['SEX']!='ALL']
```

```python
gp30_34 = gpAge[gpAge['AGE_GROUP_5']=='30_34']
```

```python
gp60_64 = gpAge[gpAge['AGE_GROUP_5']=='60_64']
```

```python
gpConcat = pd.concat([gp30_34, gp60_64], axis=0)
```

An alternative way of doing this

```python
#gpConcat = gp30_34.append(gp60_64)
```

```python
gpConcat['NUMBER_OF_PATIENTS_10000s'] = gpConcat[["NUMBER_OF_PATIENTS"]] / 10000
gpConcat = gpConcat[['SEX', 'NUMBER_OF_PATIENTS_10000s', 'AGE_GROUP_5']]
gpConcat.sample(20)
```

```python
gpConcatGroup = gpConcat.groupby(["SEX", "AGE_GROUP_5"]).sum()
gpConcatGroup = gpConcatGroup.round()
```

```python
gpConcatGroup.plot(kind="bar")
plt.show()
```

How would you quickly change this to show 90-94?

**Quick fixes to make the plot easier to read:**

Change the size

```python
gpConcatGroup.plot(kind="bar", figsize=(10,5))
plt.show()
```

Give it a title

```python
gpConcatGroup.plot(kind="bar", figsize=(10,5), title="Comparing number of patients by Age Group and Sex")
plt.show()
```

Move the legend

```python
gpConcatGroup.plot(kind="bar", figsize=(10,5), title="Comparing number of patients by Age Group and Sex").legend(bbox_to_anchor=(1, 0.5))
plt.show()
```

Change the styling

```python
import matplotlib
matplotlib.style.use('ggplot')

gpConcatGroup.plot(kind="bar", figsize=(10,5), title="Comparing number of patients by Age Group and Sex").legend(bbox_to_anchor=(1, 0.5))
plt.show()
```
