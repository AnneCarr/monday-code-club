**If we wanted to merge this Life Expectancy dataframe with the GDP dataframe - what would we need to do to the data?**

**A simple function**

Function blocks begin with the keyword def followed by the function name and parentheses ( )

Any input parameters or arguments should be placed within these parentheses. You can also define parameters inside these parentheses.

The code block within every function starts with a colon (:) and is indented.

The function will end with by returning something

```python
def upper(text):
    return text.upper()
```

Call the function by giving it some text

```python
a = "you could do this in a few different ways but it might be helpful to see a simple function"
upper(a)
```

Use apply to run the upper function on a column

```python
dfLife["Country Code"].apply(upper)
```

**Pandas also has a way of doing this**

Extract your column and assign it to a new variable

```python
s = dfLife["Country Code"]
s
```

Use str.upper to capitalise it

```python
s.str.upper()
```

Pick which way you'd like to do it and add the column back to your dataframe

```python
dfLife["Country Code"] = dfLife["Country Code"].apply(upper)
dfLife
```

**Merge the dataframes**

When combining two datasets on a common column you use merge. 

```python
from IPython.display import Image
Image(url='http://www.dofactory.com/Images/sql-joins.png')
```

As both dataframes have years in them we'll need to use suffixes to differentiate between GDP and Life Expectancy

```python
df = pd.merge(dfGdp, dfLife, how="left", on="Country Code", suffixes=("_Gdp", "_LE"))
df.sample(5)
```

## Plotting GDP and Life Expectancy

```python
df.plot.scatter("2015_Gdp", "2015_LE", figsize=(15,10))
plt.show()
```

```python
df[["2015_Gdp", "2015_LE"]].describe()
```

Plot the 50 countries with the lowest life expectancy

```python
df.nsmallest(50, "2015_LE", keep='first').plot.scatter("2015_Gdp", "2015_LE", figsize=(15,10))
plt.show()
```

Plot the 50 countries with the highest life expectancy 

```python
df.nlargest(50, "2015_LE", keep='first').plot.scatter("2015_Gdp", "2015_LE", figsize=(15,10))
plt.show()
```

## Create an average Life Expectancy column

Let's create an average column, to see how life expectancy across all countries has changed over time

```python
dfLifeAvg = dfLife.copy()
dfLifeAvg = dfLifeAvg.set_index("Country Name")
```

```python
dfLifeAvg = dfLifeAvg[[
       '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016']]
```

```python
dfLifeAvg = dfLifeAvg.transpose()
dfLifeAvg
```

Create a new column and set it to the average of the rows

```python
dfLifeAvg["avgAll"] = dfLifeAvg.mean(axis=1)
dfLifeAvg
```

Plot it

```python
dfLifeAvg.avgAll.plot.line(figsize=(10,7))
plt.show()
```
