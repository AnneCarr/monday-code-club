We only want the columns which contain the word 'TOTAL' - there are a few ways to do this, but let's try list comprehension. http://treyhunner.com/2015/12/python-list-comprehensions-now-in-color/ 

We can build it up gradually to see what's happening, firstly in a for loop:

```python
for c in dfLa.columns:
    if "TOTAL" in c:
        print(c)
```

```python
[c for c in dfLa.columns if "TOTAL" in c]
```

Then on the dataframe - copy your original dataframe and apply the list comprehension 

```python
dfLaTotals = dfLa.copy()
dfLaTotals = dfLaTotals[[c for c in dfLa.columns if "TOTAL" in c]]
dfLaTotals.head()
```

This petition 'Make it illegal for a company to require women to wear high heels at work' was debated in Parliment and has a large number of rows for us to work with.

```python
petitions_data = requests.get('https://petition.parliament.uk/petitions/129823.json')
```

To view the data use .text after your variable name

```python
petitions_data.text
```

We need to convert the requests object to a JSON object

When we import JSON data into python it becomes a dictionary (dict) object and we can use dict operations on it
https://docs.python.org/2/tutorial/datastructures.html#dictionaries

```python
petitions_dict = petitions_data.json()
petitions_dict
```

Now we are in a position to start exploring the structure of the data

```python
petitions_dict.keys()
```

A dict is made up of a key value pair like {key : value} and we can access the value using the key.

Lets explore the data value

```python
petitions_dict['data']
```

We can look at the keys within the data value

```python
petitions_dict['data'].keys()
```

Lets look at the keys within 'attributes'

```python
petitions_dict['data']['attributes'].keys()
```

Just general info there so lets have a look at the 'id' key and see what that's like

```python
petitions_dict['data']['id']
```

Well that's not much good to us at the moment 

Lets try the type

```python
petitions_dict['data']['type']
```

Mmm great it's a petition

So lets go back and explore the attributes a bit more

'signatures_by_constituency' might be something we can use - lets look at that 

```python
petitions_dict['data']['attributes']['signatures_by_constituency']
```

That's a bit more interesting - lets pull that into a dataframe and see what we can do to it

```python
const = pd.DataFrame.from_dict(petitions_dict['data']['attributes']['signatures_by_constituency'])
print('{} constituancies voted'.format(len(const)))
print(const.head())
```

Which constituency had the most votes?

```python
const.sort_values(by='signature_count',ascending=False).head()
```

We can have a look if people from other countries also voted

```python
petitions_dict['data']['attributes']['signatures_by_country']
```

Put this into a dataframe

```python
country = pd.DataFrame.from_dict(petitions_dict['data']['attributes']['signatures_by_country'])
country.sort_values(by='signature_count', ascending=False).head()
```

Lets add a column to the each dataframe called 'petition_name'

```python
const['petition_name'] = petitions_dict['data']['attributes']['action']
const.head()
```

```python
country['petition_name'] = petitions_dict['data']['attributes']['action']
country.head()
```
